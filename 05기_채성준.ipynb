# 심층 신경망

```python
import tensorflow as tf

tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()
```

## 2개의 층

### new 개념 = 은닉층

입력층과 출력층 사이에 있는 모든 층

활성화 함수가 표시되어 있음

여기서 활성화 함수란 신경망 층의 선형 방정식의 계산 값에 적용하는 함수

```python
from tensorflow import keras

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split

train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
# dense1 = 은닉층(100개의 뉴런을 가진 밀집층)
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784,))

# dense2 = 10개의 뉴런을 두어 클래스 분류(소프트맥스 함수 지정)
dense2 = keras.layers.Dense(10, activation='softmax')
```

## 심층 신경망 만들기

```python
# 앞서 만든 inputs, dense1, dense2를 클래스에 추가해서 만들어보기
model = keras.Sequential([dense1, dense2])
```

```python
# summary() = 층에 대한 유용한 정보 호출
model.summary()
```

## 층을 추가하는 다른 방법

```python
model = keras.Sequential([
    keras.layers.Dense(100, activation='sigmoid', input_shape=(784,), name='hidden'),
    keras.layers.Dense(10, activation='softmax', name='output')
], name='패션 MNIST 모델')
```

```python
model.summary()
```

```python
# add() 메서드를 호출해서 층을 추가하기

model = keras.Sequential()
model.add(keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# 모델 훈련 - compile()

model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_scaled, train_target, epochs=5)
```

## 렐루 활성화 함수

기존 시그모이드 함수: 올바른 출력을 만드는데 신속하게 대응 불가

특히 층이 많은 심층 신경망일수록 그 효과가 누적되어 학습을 더 어렵게 만든다.

그걸 개선하기 위해 다른 종류의 활성화 함수가 제안되었다.

그것이 바로 **렐루**함수

-> 마치 활성화 함수가 없던 것처럼 그냥 입력을 통과시키는데, 음수는 0으로 처리

```python
# 질문 = 교재에서 시그모이드 그래프를 보여주면서
# 올바른 출력을 만드는데 신속하게 대응한다고 하면서 그래프를 참고하라고 하는데,
# 그래프를 봐도 왜 신속하게 대응이 안되는 단점이 생기는지 이해가 되지 않는다.
# 특히 심층 신경망일수록 효과가 누적된다는데, 왜 그런지 자세하게 알고 싶다
```

```python
# Flatten 클래스 = 배치 차원 제외 나머지 입력 차원을 모두 일렬로 펼치는 기능
model = keras.Sequential()
# Flatten 층이 1차원으로 펼쳐줄 것이기에 2차원 입력 받기 (28, 28)
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# reshape() 메서드 적용 x
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
# 모델 훈련
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_scaled, train_target, epochs=5)
```

```python
# 검증 세트에서의 성능 확인
model.evaluate(val_scaled, val_target)
```

## 옵티마이저

옵티마이저?

케라스는 다양한 종류의 경사 하강법 알고리즘을 제공

이들을 옵티마이저라고도 부른다!

```python
# SGD 옵티마이저를 사용하기 위해 'sgd'로 지정
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# 위의 코드와 정확히 동일
sgd = keras.optimizers.SGD()
model.compile(optimizer=sgd, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# SGD 클래스의 학습률 기본값은 0.01, 하지만 0.1로 재지정
sgd = keras.optimizers.SGD(learning_rate=0.1)
```

#### 모멘텀 최적화

기본 경사 하강법 옵티마이저 = 모두 SGD 클래스에서 제공

그래서 momontum 매개변수의 기본값은 0

이를 0보다 큰 값으로 지정하면,

마치 이전의 그레디언트를 가속도처럼 사용하는 모멘텀 최적화를 사용

```python
# nestreov 매개변수를 기본값 False에서 True로 바꾸면 네스테로프 모멘텀 최적화 사용
sgd = keras.optimizers.SGD(momentum=0.9, nesterov=True)
```

적응적 학습률 = 모델이 최적점에 가까이 갈수록 학습률을 낮출 수 있는데,

이렇게 하면 안정적으로 최적점에 수렴할 가능성이 높다!

그리고 이런 학습률을 **적응적 학습률**이라고 한다!!!

```python
# Adagrad = 적응적 학습률의 대표적 옵티마이저

adagrad = keras.optimizers.Adagrad()
model.compile(optimizer=adagrad, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# RMSprop = 적응적 학습률의 대표적 옵티마이저 2

rmsprop = keras.optimizers.RMSprop()
model.compile(optimizer=rmsprop, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

learning_rate 매개변수의 기본값으로 모두 0.001을 사용
