### 1번 제출 답안

```python
shopping_review = [] # 비어있는 리스트 생성
#for 문으로 빈 리스트에 append 형식으로 채우기

for line in data:
    temp = line.split('\t')
    temp_dict= {}
    temp_dict['score'] = temp[0] # score 키 생성 - 첫 번째 줄에 있는 게 평점 = 0
    temp_dict['review'] = temp[1].replace('\n','') # review 키 생성 - 두 번째 줄에 있는 게 리뷰 = 1, '\n'은 빈칸으로 변경
    shopping_review.append(temp_dict) # for문 안에서 반복적으로 값 저장됨
```

## 2. 점수 5, 4, 3, 2, 1에 대해 각각

* 리뷰 개수
* 평균 글자수 (소수점 둘째 자리까지) 를 출력하시오.

### 문제 2번 답안 - 1

각 점수별 변수와 점수별 댓글 개수 변수를 하나하나 만들어서 작성했다.

점수는 정확히 나왔으나, 너무 난잡하고 비효율적이라는 생각이 들었다.

```python
# 각 점수별 변수 생성
rate_1 = 0
rate_2 = 0
rate_3 = 0
rate_4 = 0
rate_5 = 0

for line in shopping_review:

    if line['score'] == '1':    # 만약 line['score'] 값이 '1'이라면 (숫자 1이 아닌 문자열 1)
        rate_1 = rate_1 + 1     # 점수 1 변수에 카운트 1

    elif line['score'] == '2':    # 마찬가지
        rate_2 = rate_2 + 1

    elif line['score'] == '3':
        rate_3 = rate_3 + 1

    elif line['score'] == '4':
        rate_4 = rate_4 + 1

    else: # 나머지는 차피 5일테니까 else 사용
        rate_5 = rate_5 + 1

print(f'점수 1 댓글 수 : {rate_1}')
print(f'점수 2 댓글 수 : {rate_2}')
print(f'점수 3 댓글 수 : {rate_3}')
print(f'점수 4 댓글 수 : {rate_4}')
print(f'점수 5 댓글 수 : {rate_5}')
```

```python
# 각 점수별 총 글자수 변수 생성
total_1 = 0
total_2 = 0
total_3 = 0
total_4 = 0
total_5 = 0

for line in shopping_review:
        if line['score'] == '1':
            total_1 = total_1 + (len(line['review']))

        elif line['score'] == '2':
            total_2 = total_2 + (len(line['review']))

        elif line['score'] == '3':
            total_3 = total_3 + (len(line['review']))

        elif line['score'] == '4':
            total_4 = total_4 + (len(line['review']))

        else:
            total_5 = total_5 + (len(line['review']))

print(f"점수 1 댓글 수 평균: {total_1/rate_1:.2f}" if rate_1 > 0 else "점수 1 댓글 수 평균: 리뷰 없음")
print(f"점수 2 댓글 수 평균: {total_2/rate_2:.2f}" if rate_2 > 0 else "점수 2 댓글 수 평균: 리뷰 없음")
print(f"점수 3 댓글 수 평균: {total_3/rate_3:.2f}" if rate_3 > 0 else "점수 3 댓글 수 평균: 리뷰 없음")
print(f"점수 4 댓글 수 평균: {total_4/rate_4:.2f}" if rate_4 > 0 else "점수 4 댓글 수 평균: 리뷰 없음")
print(f"점수 5 댓글 수 평균: {total_5/rate_5:.2f}" if rate_5 > 0 else "점수 5 댓글 수 평균: 리뷰 없음")
```

### 문제 2번 답안 - 2

아예 1~5점까지의 데이터를 담을 '리스트'를 설정해버렸다

기존의 나처럼 10개 이상의 변수를 굳이 설정할 필요가 없을 정도로 말이다

```python
# 초기화: 1점~5점까지의 데이터를 담을 리스트 설정 (0으로 초기화)
# index 0은 1점, index 1은 2점 ... index 4는 5점을 의미함
counts = [0, 0, 0, 0, 0]
totals = [0, 0, 0, 0, 0]

# 데이터 처리 (한 번의 루프로 개수와 글자수 합계를 동시에 계산)
for line in shopping_review:
    # 문자열인 점수를 숫자로 변환
    score_num = int(line['score'])

    # 리스트의 인덱스는 0부터 시작하므로 점수에서 1을 뺌
    index = score_num - 1

    # 해당 점수 위치의 값들을 업데이트
    counts[index] = counts[index] + 1
    totals[index] = totals[index] + len(line['review'])

# 결과 출력 (반복문을 사용하여 5점에서 1점 순서로 출력)
for i in range(4, -1, -1):  # 4(5점)부터 0(1점)까지 거꾸로 반복
    score = i + 1
    print(f"점수 {score} 댓글 수 : {counts[i]}")

    if counts[i] > 0:
        avg = totals[i] / counts[i]
        print(f"점수 {score} 평균 글자수 : {avg:.2f}")
    else:
        print(f"점수 {score} 평균 글자수 : 리뷰 없음")
    print() # 한 줄 띄우기
```

## 3. 검색어를 입력받고 해당 키워드가 포함된 리뷰에 대해

* 검색 결과 리뷰수
* 리뷰 내용 5개 출력
* 점수 평균(hint : 점수를 숫자 형식으로 다루어야 함. 출력은 소수점 둘째 자리까지)을 출력하시오.
* (hint : 검색 결과를 별도의 리스트로 구성하는 방법을 고려할 수 있음. 필수 요건은 아님)
* 참고) 검색 결과가 없을 때는 내용 출력이나 평균을 구할 수 없으므로 검색 결과가 나오는 키워드를 사용할 것

### 3번 문제 제출 답안

```python
# 검색어 입력받기용 변수
search = input("검색할 키워드를 입력하세요: ")

# 키워드가 포함된 리뷰 필터링 (Hint: 별도의 리스트로 구성)
keyword = [r for r in shopping_review if search in r['review']]

# 검색 결과 리뷰 수
result = len(keyword)
print(f"검색 결과 리뷰 수: {result}개")

# 검색 결과가 있을 경우에만 실행
if result > 0:
    # 5개보다 적을 수도 있으니, 둘 중 작은 값을 반복 횟수로 설정
    count = 5
    if result < 5:
        count = result

    for i in range(count):
        print(i + 1, ".", keyword[i]["review"])

    # 점수 평균 계산 (합계를 구한 뒤 개수로 나누기)
    score = 0
    for review in keyword:
        score = score + int(review["score"]) # 점수를 숫자 형식으로 더함 (문자열을 정수로 변환)

    average_score = score / result

    # 평균 출력 (소수점 둘째 자리까지 출력하는 format 함수 사용)
    print("\n점수 평균:", format(average_score, ".2f"))
else:
    print("해당 키워드가 포함된 리뷰가 없습니다.")
```

## 4. 검색어를 입력받고 해당 키워드가 포함된 리뷰들을 search_result.txt 파일에 저장하시오.
* 점수와 리뷰 내용

```python
# 검색어 입력받기용 변수
keyword = input("검색할 키워드를 입력하세요: ")

with open("search_result.txt", "w") as f:
    for line in shopping_review:
        if keyword in line['review']:
            f.write(f"{line['score']}\t{line['review']}\n")

print("저장 완료")
```
